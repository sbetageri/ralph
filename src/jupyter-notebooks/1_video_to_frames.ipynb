## File Preparation

### 1) Test & Create Required Functions To Process Files

#### Load Required Modules

```python
import os
from os import listdir
from os.path import isfile, join, isdir
import cv2
```

#### Variable Initialization

```python
folders = ['./trainval','./test']
tfolders = ['./ttrainval','./ttest']
```

#### Directory and File Utilities

```python
def getSubDirs(dirpath):
    files = [f for f in listdir(dirpath) if not f.startswith('.') and isdir(join(dirpath,f))]
    return files

len(getSubDirs('./test'))
```

```python
def getMP4Files(dirpath):
    files = [f.replace('.mp4','') for f in listdir(dirpath) if not f.startswith('.') and f.endswith('.mp4') and isfile(join(dirpath,f))]
    return files

len(getMP4Files('./trainval/0bop3D7SdDM'))
```

#### Video Processor

```python
def videoToFrames(srcPath, destPath, subdir, filename):
    cap = cv2.VideoCapture(join(srcPath, subdir, filename + '.mp4'))
    if not os.path.exists(join(destPath, subdir)):
        os.makedirs(join(destPath, subdir))
    if not os.path.exists(join(destPath, subdir, filename)):
        os.makedirs(join(destPath, subdir, filename))
    success, image = cap.read()
    count = 0
    while success:
        #print(str(count) + ": " + str(success))
        newPath = join(destPath, subdir, filename, str(count) + '.jpg')
        #print("\t" + newPath)
        cv2.imwrite(newPath, image)     # save frame as JPEG file
        success, image = cap.read()
        count += 1
    print("\t%s\t%s\t%s" % (subdir,filename,count))
        
videoToFrames('./trainval', './ttrainval/', '0bop3D7SdDM', '50001')
```

```python
for f, folder in enumerate(folders, start=0):
    srcPath = folder
    destPath = tfolders[f]
    print(srcPath+"\t => \t"+destPath)
    print("\t%s\t%s\t%s" % ("SUBDIRECTORY","FILE","# FRAMES"))
    subdirs = getSubDirs(srcPath)
    for subdir in subdirs:
        filenames = getMP4Files(join(srcPath, subdir))
        for filename in filenames:
            videoToFrames(srcPath, destPath, subdir, filename)
```
